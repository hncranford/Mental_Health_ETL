```python
#perform imports
import os
import pandas as pd
import psycopg2
from sqlalchemy import create_engine
```

```python
#create data frame for master.csv data
masterDF = pd.read_csv("master.csv")
masterDF
```

```python
#filter master csv file data frame for the year 2016
year2016DF = masterDF[masterDF['year']==2016]
year2016DF
```

```python
#filter the 2016 master data for Croatia
croatiaDF = year2016DF[year2016DF['country']=='Croatia']
croatiaDF
```

```python
#create data frame for human resources csv file
hrDF =pd.read_csv('Human Resources.csv')
hrDF
```

```python
#filter hr data frame for the year 2016
hrYearDF = hrDF[hrDF['Year']==2016]
hrYearDF
```

```python
#rename Country column to country in order to perform a merge on country
hrYearDF = hrYearDF.rename(columns={'Country': 'country'})
hrYearDF
```

```python
#filter hr 2016 data for Croatia
hrCroatiaDF = hrYearDF[hrYearDF['country']=='Croatia']
hrCroatiaDF
```

```python
#merge croatia 2016 data frame with hr croatia 2016 data
croatia2016DF = croatiaDF.merge(hrCroatiaDF, on='country')
croatia2016DF
```

```python
#create a data frame for crude suicide rates csv file
crudeRatesDF = pd.read_csv('Crude suicide rates.csv')
crudeRatesDF
```

```python
#filter out crude rates data frame for croatia
croatiaCrudeRatesDF = crudeRatesDF[crudeRatesDF["Country"]=="Croatia"]
croatiaCrudeRatesDF
```

```python
#rename Country column to country to perform a merge
croatiaCrudeRatesDF = croatiaCrudeRatesDF.rename(columns={'Country': 'country'})
croatiaCrudeRatesDF
```

```python
#merge croatia 2016 data frame with croatia rate 2016 data frame
croatiaRate2016DF = croatia2016DF.merge(croatiaCrudeRatesDF, on='country')
croatiaRate2016DF
```

```python
#create data frame for facilities csv file
facilitiesDF = pd.read_csv('Facilities.csv')
facilitiesDF
```

```python
#filter out facilites data frame for croatia
croatiaFacilitiesDF = facilitiesDF[facilitiesDF['Country']=='Croatia']
croatiaFacilitiesDF
```

```python
#rename croatia facilities data frame Country column to country to perform a merge
croatiaFacilitiesDF = croatiaFacilitiesDF.rename(columns={'Country': 'country'})
croatiaFacilitiesDF
```

```python
#merge croatia rate 2016 data frame with croatia facilities data frame on country
croatiaFinal2016DF = croatiaRate2016DF.merge(croatiaFacilitiesDF, on='country')
croatiaFinal2016DF.head()
```

```python
#view columns in ellipse
print(croatiaFinal2016DF.columns)
```

```python
#drop unnecessary columns
croatiaFinal2016DFClean = croatiaFinal2016DF.drop(['country-year', 'population', 'HDI for year', 'Year_y', 'residential_facilities',
                            'suicides_no', ' gdp_for_year ($) ', 'gdp_per_capita ($)', 'Year_x', 'Sex'], axis=1)
croatiaFinal2016DFClean.head()
```

```python
#rename columns to match postgreSQL table
croatiaFinal2016DFClean = croatiaFinal2016DFClean.rename(columns={'suicides/100k pop' : 'suicides_pop',
                                                                  'Psychiatrists' : 'psychiatrists',
                                                                  'Nurses' : 'nurses',
                                                                  'Social_workers' : 'social_workers',
                                                                  'Psychologists' : 'psychologists',
                                                                  'Mental _hospitals' : 'mental_hospitals',
                                                                  'outpatient _facilities' : 'outpatient_facilities',
                                                                  'day _treatment' : 'day_treatment',
                                                                  ' 80_above' : 'eightyplus', ' 70to79' : 'seventies',
                                                                 ' 60to69 ' : 'sixties', ' 50to59 ' : 'fifties',
                                                                 ' 40to49' : 'forties', ' 30to39' : 'thirties',
                                                                 ' 20to29' : 'twenties', ' 10to19' : 'teens'})
croatiaFinal2016DFClean.head()
```

```python
#view columns in ellipse
print(croatiaFinal2016DFClean.columns)
```

```python
#create a connection with postgresql
rds_connection_string = "postgres:postgres@localhost:5432/croatia"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#find table names in postgresql
engine.table_names()
```

```python
#export data from pandas to postgresql
croatiaFinal2016DFClean.to_sql(name='croatia', con=engine, if_exists='append', index=False)
```

```python
#confirm table was created
pd.read_sql_query('select * from croatia', con=engine).head()
```
